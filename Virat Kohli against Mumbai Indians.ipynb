```python
from pandas import *
df = read_csv(r'C:\Users\harsi\OneDrive\Desktop\WPL Datasets\deliveries.csv')
df = df.set_index('match_id')
```

```python
team = ['Mumbai Indians', 'Royal Challengers Bangalore']
MIvsRCB = df.loc[df['batting_team'].isin(team) & df['bowling_team'].isin(team)]
Kohli_dismissals = MIvsRCB.loc[(MIvsRCB['player_dismissed'] == 'V Kohli')]
Kohli_dismissals = Kohli_dismissals[['bowler', 'inning']]
Kohli_runs = MIvsRCB.loc[(MIvsRCB['batsman'] == 'V Kohli')].groupby('match_id').agg({'batsman_runs':['sum']})
Kohli_runs.columns = ['kohli_runs']
balls = MIvsRCB.loc[(MIvsRCB['batsman'] == 'V Kohli') & (MIvsRCB['noball_runs'] == 0) & (MIvsRCB['wide_runs'] == 0)]
balls = balls.groupby('match_id').count()
balls = balls['ball']
result = concat([Kohli_dismissals, Kohli_runs, balls], axis = 1).fillna(0)
result['SR'] = (result['kohli_runs']*100/result['ball']).astype(int)
result.to_csv(r'C:\Users\harsi\OneDrive\Desktop\WPL Datasets\KohlivsMI.csv')
result
```
