```python
from pandas import *
```

```python
df = read_csv(r'C:\Users\harsi\OneDrive\Desktop\WPL Datasets\deliveries.csv')
print(df)
```

```python
QdK = df.loc[(df['batsman'] == 'Q de Kock') & (df['player_dismissed'] == 'Q de Kock')]
print(QdK)
```

```python
QdK_wickets = QdK[['match_id', 'bowler', 'player_dismissed']].dropna()
print(QdK_wickets)
```

```python
bowlers = QdK_wickets.groupby('bowler').size()
print(bowlers)
```

```python
bowlers.to_csv(r'C:\Users\harsi\OneDrive\Desktop\WPL Datasets\bowlers.csv')
```
