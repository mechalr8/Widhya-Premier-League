```python
from pandas import *
```

```python
df = read_csv(r'C:\Users\Asus\Desktop\WPL Datasets\Bangalore vs Kolkata deliveries.csv')
print(df)
```

```python
kohlifours = df.loc[(df['batsman'] == 'V Kohli') & (df['batsman_runs'] == 4)]
print(kohlifours)
```

```python
matches = kohlifours.groupby('match_id').size().count()
print(matches)
```

```python
index = kohlifours.index
fours = len(index)
print(fours)
```

```python
avg = fours/matches
print(avg)
```
